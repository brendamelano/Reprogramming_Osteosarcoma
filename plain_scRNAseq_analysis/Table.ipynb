```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a DataFrame with the provided information
data = {
  'Sample': ['OS152', 'OS384', 'OS742'],
  'Amplification': ['MYC', 'VEGFA', 'MYC and CDK4'],
  'Source': ['Metastatic', 'Resection',  'Diagnostic']
}

df = pd.DataFrame(data)

# Plot table and remove axis
fig, ax = plt.subplots(figsize=(6, 2))  # Adjust the size as necessary
ax.axis('tight')
ax.axis('off')
ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

plt.show()
```
